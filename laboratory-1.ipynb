```python
import cv2
from matplotlib import pyplot as plt 
import numpy as np
```

```python
def convertImage(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
def plotImage (image) :
    plt.axis('off')
    plt.imshow(convertImage(image))
    plt.show()
```

```python
R, G, B = (2, 1, 0)
```

```python
def changeColor(imageColor, option):

    image = imageColor.copy()

    if option == 1:
        image[:,:,G] = 0
        image[:,:,R] = 0

    elif option == 2:
        image[:,:,B] = 0
        image[:,:,R] = 0
    
    elif option == 3:
        image[:,:,G] = 0
        image[:,:,B] = 0

    elif option == 10:
        image[:,:,B] = 0

    elif option == 20:
        image[:,:,R] = 0

    elif option == 30:
        image[:,:,G] = 0

    return image
```

```python
image = cv2.imread("test.png")

f, axarr = plt.subplots(2,3)
axarr[0, 0].imshow(convertImage(changeColor(image, 1)))
axarr[0, 1].imshow(convertImage(changeColor(image, 2)))
axarr[0, 2].imshow(convertImage(changeColor(image, 3)))
axarr[1, 0].imshow(convertImage(changeColor(image, 10)))
axarr[1, 1].imshow(convertImage(changeColor(image, 20)))
axarr[1, 2].imshow(convertImage(changeColor(image, 30)))

axarr[0, 0].axis('off')
axarr[0, 1].axis('off')
axarr[0, 2].axis('off')
axarr[1, 0].axis('off')
axarr[1, 1].axis('off')
axarr[1, 2].axis('off')
```

```python
def rebuildImage(redImage, greenImage, blueImage):
    image = redImage.copy()

    for y in range(len(image)):
        for x in range(len(image[0])):        
            red = redImage[y][x][2]
            green = greenImage[y][x][1]
            blue = blueImage[y][x][0]

            image[y][x] = [blue, green, red]

    return image
```

```python
redImage1 = cv2.imread("imagen1/imagen1_salida_gray_rojo.jpg")
greenImage1 = cv2.imread("imagen1/imagen1_salida_gray_verde.jpg")
blueImage1 = cv2.imread("imagen1/imagen1_salida_gray_azul.jpg")

image1 =  rebuildImage(redImage1, greenImage1, blueImage1)


redImage2 = cv2.imread("imagen2/imagen2_salida_gray_rojo.jpg")
greenImage2 = cv2.imread("imagen2/imagen2_salida_gray_verde.jpg")
blueImage2 = cv2.imread("imagen2/imagen2_salida_gray_azul.jpg")

image2 =  rebuildImage(redImage2, greenImage2, blueImage2)


redImage3 = cv2.imread("imagen3/perro_salida_gray_rojo.jpg")
greenImage3 = cv2.imread("imagen3/perro_salida_gray_verde.jpg")
blueImage3 = cv2.imread("imagen3/perro_salida_gray_azul.jpg")

image3 =  rebuildImage(redImage3, greenImage3, blueImage3)

plotImage(image1)
plotImage(image2)
plotImage(image3)
```

```python
def grayImage(image) :
    redImage = image.copy()
    greenImage = image.copy()
    blueImage = image.copy()

    for y in range(len(image)):
        for x in range(len(image[0])):
            pixel = image[y][x]

            b, g, r = pixel

            redImage[y][x] = [r, r, r]
            greenImage[y][x] = [g, g, g]
            blueImage[y][x] = [b, b, b]

    return redImage, greenImage, blueImage
```

```python
colorImage = cv2.imread("test.png")


red, green, blue = grayImage(colorImage)

print("Rojo")
plotImage(red)
print("Verde")
plotImage(green)
print("Azul")
plotImage(blue)


imageBuilded = rebuildImage(red, green, blue)

plotImage(imageBuilded)
```

```python
def histogram(image):
  newImage = image.copy()

  blueData = []
  greenData = []
  redData = []
  grayScale = []

  for y in range(len(image)):
    for x in range(len(image[0])):   

      pixel = newImage[y][x]

      blueData.append(pixel[B])
      greenData.append(pixel[G])
      redData.append(pixel[R])
      grayScale.append(np.average(pixel))


  plt.title('Rojo')
  redHistogram, bins = np.histogram(redData, bins=range(256))
  plt.plot(redHistogram)
  plt.show()

  plt.title('Verde')
  greenHistogram, bins = np.histogram(greenData, bins=range(256))
  plt.plot(greenHistogram)
  plt.show()


  plt.title('Azul')
  blueHistogram, bins = np.histogram(blueData, bins=range(256))
  plt.plot(blueHistogram)
  plt.show()


  plt.title('Escala de Grises')
  grayHistogram, bins = np.histogram(grayScale, bins=range(256))
  plt.plot(grayHistogram)
  plt.show()
```

```python
image = cv2.imread('test.png')
plotImage(image)
histogram(image)
```
