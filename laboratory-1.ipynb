```python
import cv2
from matplotlib import pyplot as plt 
import numpy as np
```

```python
def convertImage(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
def plotImage (image) :
    plt.axis('off')
    plt.imshow(convertImage(image))
    plt.show()
```

```python
R, G, B = (2, 1, 0)
```

```python
def changeColor(imageColor, option):

    image = imageColor.copy()

    if option == 1:
        image[:,:,G] = 0
        image[:,:,R] = 0

    elif option == 2:
        image[:,:,B] = 0
        image[:,:,R] = 0
    
    elif option == 3:
        image[:,:,G] = 0
        image[:,:,B] = 0

    elif option == 10:
        image[:,:,B] = 0

    elif option == 20:
        image[:,:,R] = 0

    elif option == 30:
        image[:,:,G] = 0

    return image
```

```python
image = cv2.imread("test.png")

f, axarr = plt.subplots(2,3)
axarr[0, 0].imshow(convertImage(changeColor(image, 1)))
axarr[0, 1].imshow(convertImage(changeColor(image, 2)))
axarr[0, 2].imshow(convertImage(changeColor(image, 3)))
axarr[1, 0].imshow(convertImage(changeColor(image, 10)))
axarr[1, 1].imshow(convertImage(changeColor(image, 20)))
axarr[1, 2].imshow(convertImage(changeColor(image, 30)))

axarr[0, 0].axis('off')
axarr[0, 1].axis('off')
axarr[0, 2].axis('off')
axarr[1, 0].axis('off')
axarr[1, 1].axis('off')
axarr[1, 2].axis('off')
```

```python
def rebuildImage(redImage, greenImage, blueImage):
    image = redImage.copy()

    for y in range(len(image)):
        for x in range(len(image[0])):        
            red = redImage[y][x][2]
            green = greenImage[y][x][1]
            blue = blueImage[y][x][0]

            image[y][x] = [blue, green, red]

    return image
```

```python
redImage1 = cv2.imread("imagen2/imagen2_salida_gray_rojo.jpg")
greenImage1 = cv2.imread("imagen2/imagen2_salida_gray_verde.jpg")
blueImage1 = cv2.imread("imagen2/imagen2_salida_gray_azul.jpg")

image1 =  rebuildImage(redImage1, greenImage1, blueImage1)


redImage2 = cv2.imread("imagen2/imagen2_salida_gray_rojo.jpg")
greenImage2 = cv2.imread("imagen2/imagen2_salida_gray_verde.jpg")
blueImage2 = cv2.imread("imagen2/imagen2_salida_gray_azul.jpg")

image2 =  rebuildImage(redImage2, greenImage2, blueImage2)


redImage2 = cv2.imread("imagen2/imagen2_salida_gray_rojo.jpg")
greenImage2 = cv2.imread("imagen2/imagen2_salida_gray_verde.jpg")
blueImage2 = cv2.imread("imagen2/imagen2_salida_gray_azul.jpg")

image2 =  rebuildImage(redImage2, greenImage2, blueImage2)
```
